## Q. Basic music theory

### Bits

This is list of the bits duration:

* Semibreve : 4x
* Minim : 2x
* Crotchet : 1x
* Quaver : $\frac{1}{2}x$
* Semiquaver : $\frac{1}{4}x$
* Demisemiquaver : $\frac{1}{8}x$

Where x is base duration, based on type of music it is a number usually between 40 and 160 and in Bit Per Minute (BPM) units.

### Notes

As you may know, music has 7 main notes in different harmonics. These notes are:

* A (La) = $2^{\frac{0}{12}}F_{b}$
* A# (La diesis/Si bemolle) = $2^{\frac{1}{12}}F_{b}$
* B (Si) = $2^{\frac{2}{12}}F_{b}$
* C (Do) = $2^{\frac{3}{12}}F_{b}$
* C# (Do diesis/Re bemolle) = $2^{\frac{4}{12}}F_{b}$
* D (Re) = $2^{\frac{5}{12}}F_{b}$
* Eb (Mi bemolle/Re diesis) = $2^{\frac{6}{12}}F_{b}$
* E (Mi) = $2^{\frac{7}{12}}F_{b}$
* F (Fa) = $2^{\frac{8}{12}}F_{b}$
* F# (Fa diesis/Sol bemolle) = $2^{\frac{9}{12}}F_{b}$
* G (Sol) = $2^{\frac{10}{12}}F_{b}$
* G# (Sol diesis/La bemolle) = $2^{\frac{11}{12}}F_{b}$

Why are they 12? just a historical notation. $F_b$ is also base frequency, for A4 it is 440Hz. for Ak it is equal to $2^{k-1}\times 55 Hz$

* These 12 half-steps form 6 full steps, this 6 steps are called a full Octave. 
* Each octave frequencies are 2x higher than previous Octave and $\frac{1}{2}$ lower than next Octave.

```python
notes_base = 2**(np.arange(12)/12)*27.5
notes_duration = np.array([3200, 1600, 800, 400, 200, 100])
notes_ann = ['A', 'A#', 'B', 'C', 'C#', 'D', 'Eb', 'E', 'F', 'F#', 'G', 'G#']
cnt = 1

def sin_wave(f, n, fs):
    x = np.linspace(0, 2*np.pi, n)
    xp = np.linspace(0, -1*(n*30/fs), n)
    y = np.sin(x*f*(n/fs))*np.exp(xp)
    z = np.zeros([n, 2])
    z[:, 0] = y
    z[:, 1] = y
    return z

def play_note(note_id, octave, dur, fs):
    global cnt
    y = sin_wave(notes_base[note_id]*2**octave, int(notes_duration[dur]*(fs/1000)), fs)
    plt.subplot(8, 2, cnt)
    cnt += 1
    plt.plot(y)
    plt.title('{}{} is playing '.format(notes_ann[note_id], octave))
    sd.play(y, fs)
    sd.wait()
    return 

def put_note(note_id, octave, dur, fs):
    y = sin_wave(notes_base[note_id]*2**octave, int(notes_duration[dur]*(fs/1000)), fs)
    return y

def get_music(music_notes, fs):
    m = []
    for item in music_notes:
        y = put_note(item[0], item[1], item[2], fs)
        m.append(y)
    m = np.concatenate(m, 0)
    return m

fs1 = 44100
music = [[8, 5, 3], [10, 5, 4], [10, 5, 4], [10, 5, 3], [10, 5, 3], [10, 5, 3], [10, 5, 3], [10, 5, 3], [10, 5, 3], 
        [11, 5, 4], [4, 5, 4], [11, 5, 4], [4, 5, 4], [11, 5, 4], [4, 5, 4], [11, 5, 4], [4, 5, 4], [3, 5, 3], 
        [8, 5, 4], [8, 5, 4], [8, 5, 3], [8, 5, 3], [8, 5, 3], [8, 5, 3], [8, 5, 3], [8, 5, 3], 
        [10, 5, 4], [3, 5, 4], [10, 5, 4], [3, 5, 4], [10, 5, 4], [3, 5, 4], [10, 5, 4], [3, 5, 4], [1, 5, 3]]

# plt.figure(figsize=(30, 30))
# for item in music[:16]:
#     play_note(item[0], item[1], item[2], fs1)
# plt.show()

y = get_music(music, fs1)
sd.play(y, fs1)
sd.wait()
```

## 1. Record and load audio data

In this phase, music/audio should be recorded or loaded from mp3/wav file. here is an example:

```python
import pydub
import numpy as np
import sounddevice as sd
from matplotlib import pyplot as plt
from scipy.io.wavfile import read, write


class Music: # static class for music I/O standard
    
    @staticmethod
    def plotter(x=None, y=None, title="", xlabel="", ylabel="", figsize=(20, 12), subplt=(1, 1, 1)):
        """
            East plotter
        """
        if x is None:
            x = np.arange(y.shape[0])
        plt.figure(figsize=figsize)
        plt.subplot(subplt)
        plt.plot(x, y)
        
        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.show()
        return
    
    @staticmethod
    def frequencies(x, fs):
        """
            Audio vector to frequencies with FFT ratio
        """
        Fd = np.fft.fft(x)
        Fx = np.linspace(0, fs, max(x.shape))
        return Fd, Fx
    
    @staticmethod
    def read_mp3(filename, normalized=False):
        """
            MP3 to numpy array
        """
        a = pydub.AudioSegment.from_mp3(filename)
        y = np.array(a.get_array_of_samples())
        if a.channels == 2:
            y = y.reshape((-1, 2))
        if normalized:
            return a.frame_rate, np.float32(y) / 2**15
        else:
            return a.frame_rate, y
    
    @staticmethod
    def write_mp3(filename, Fs, x, normalized=False):
        """ 
            numpy array to MP3 
        """
        channels = 2 if (x.ndim == 2 and x.shape[1] == 2) else 1
        if normalized:  
            y = np.int16(x * 2 ** 15)
        else:
            y = np.int16(x)
        song = pydub.AudioSegment(y.tobytes(), frame_rate=Fs, sample_width=2, channels=channels)
        song.export(filename, format="mp3", bitrate="320k")
        return


fs = 44100  # Sample rate
seconds = 1  # Duration of recording

fs1, data1 = Music.read_mp3('Capriccio n.24 Op.1 in A minor.mp3')
# myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=2)
# sd.wait()  # Wait until recording is finished
Music.write_mp3('Cap no.24 en A minor.mp3', fs1, data1[:1000000, :], normalized=False)
# sd.play(data1[70000:200000], fs1*1.18) # x1.18 fast
# sd.wait()
# sd.play(myrecording, fs*0.67) # x0.67 slow
# sd.wait()
# write('musics/output.wav', fs, myrecording) 
print('Rate = {}, Music data shape = {}'.format(fs1, data1.shape))
```

Filter example:

```python
def freqplot(y, fs, N, t1="The signal", t2="Frequency"):

    x = np.linspace(0, 2*np.pi, y.shape[0]) # Zero to 2*pi, one period
    f = np.fft.fft(y)
    p = np.linspace(0, fs, y.shape[0])

    plt.figure(figsize=(30, 10))
    
    plt.subplot(2, 1, 1)
    plt.plot(x, y)
    plt.grid(True)
    plt.title(t1)
    
    plt.subplot(2, 1, 2)
    plt.plot(p[:N], np.abs(f[:N]))
    plt.grid(True)
    plt.title(t2)
    plt.show()
    
    
fs = 44100  # Sample rate
fs1, data1 = Music.read_mp3('musics/RV521 Concerto In A For 2 Violins.mp3')
```

```python
x = np.linspace(-7.85, 7.86, 101)
lfilter1 = np.sin(x)/x
nm = np.sum(lfilter1)
lfilter1 /= nm

x = np.linspace(-7.85, 7.86, 601)
lfilter2 = np.sin(x)/x
nm = np.sum(lfilter2)
lfilter2 /= nm

d1 = np.convolve(data1[:, 1], lfilter1)
d2 = np.convolve(data1[:, 1], lfilter2)

freqplot(lfilter1, fs, 18, "Lowpass filter 1")
freqplot(lfilter2, fs, 18, "Lowpass filter 2")
```

```python
y = data1[75000:85000, 1]
y1 = d1[75050:85050]
y2 = d2[75300:85300]

freqplot(y, fs, 400)
freqplot(y1, fs, 400, "First low-pass filtered")
freqplot(y2, fs, 400, "Second low-pass filtered")
freqplot(y1 - y2, fs, 400, "Mixed bandpass filtered")
```

## 4. Time domain decomposition

Here we propose an idea. Waveform of instruments like violin is almost constant for each frequency. so if you knew that the audio contains violin with which amplitude and frequency, you'll be able to subtract that from your audio and get notes for other instruments.

This is somehow called: Discrete wavelet transform.

Our idea is for two instruments: let $v(f_1)$ as violin sound time form at frequency $f_1$ and $p(f_2)$ as piano sound at frequency $f_2$, the input audio should be $\alpha v(f_1) + \beta p(f_2)$ at a single bit, actually they have unique harmonic ratios; so we can write two equations for harmonic ratios and input signals and get alpha and beta coefficients. 

This idea is applicable to more than two instruments, but all you need is to form new equations from harmonic series! as a hint, it could be solved with convex optimization methods, first start with simple approximations.
